```python
import cv_packages as draw
import cv2
import numpy as np
from matplotlib import pyplot as plt

draw.set_figsize((4, 6))
```

```python
image = cv2.imread('labl.jpg')
draw.show_image(image, 'Original Image')
```

```python
gauss_image1 = cv2.GaussianBlur(image, ksize=(55, 55), sigmaX=0, sigmaY=0)
gauss_image2 = cv2.GaussianBlur(image, ksize=(55, 55), sigmaX=100, sigmaY=100)
draw.show_image(image)
draw.show_image(gauss_image1, 'Gauss Image1')
draw.show_image(gauss_image2, 'Gauss Image2')
```

```python
img_blur1 = cv2.blur(image, (7,9)) # cv2.blur(src,ksize,dst,anchor,borderType)
img_blur2 = cv2.blur(image, (55, 55))
draw.show_image(image)
draw.show_image(img_blur1, 'Blur Image1')
draw.show_image(img_blur2, 'Blur Image2')
```

```python
pepper_image = draw.add_pepper_noisy(image, 100000) # cv2.medianBlur(src,ksize,dst)
img_medianblur = cv2.medianBlur(pepper_image, 5)
draw.show_image(pepper_image, 'Pepper Image')
draw.show_image(img_medianblur, 'MedianBlur Image')
```

```python
#cv2.bilateralFilter(src,d,sigmaColor,sigmaSpace,borderType)
img_bilateral1 = cv2.bilateralFilter(image, 0, 0.2, 40)
img_bilateral2 = cv2.bilateralFilter(image, 0, 0.2, 100)
draw.show_image(image)
draw.show_image(img_bilateral1, 'Bilateral Image1')
draw.show_image(img_bilateral2, 'Bilateral Image2')
```

```python
# cv2.ximgproc.guidedFilter(guide,src,radius,eps,dDepth)
guide_filtered_image1 = cv2.ximgproc.guidedFilter(guide=image, src=image, radius=9, eps=75)
guide_filtered_image2 = cv2.ximgproc.guidedFilter(guide=image, src=image, radius=111, eps=75)
draw.show_image(image)
draw.show_image(guide_filtered_image1, 'GuideFiltered Image1')
draw.show_image(guide_filtered_image2, 'GuideFiltered Image2')
```
